```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
def E_field_point_charges(X, Y, charges):
    Ex, Ey = np.zeros_like(X), np.zeros_like(Y)
    for charge in charges:
        dx = X - charge["pos"][0]
        dy = Y - charge["pos"][1]
        r2 = dx**2 + dy**2
        r3 = np.maximum(r2**1.5, 1e-10)  # Avoid division by zero
        Ex += charge["q"] * dx / r3
        Ey += charge["q"] * dy / r3
    return Ex, Ey
```

```python
# Simple model to show electric field from a charged rod along x.
rod_length = 5
num_charges = 100
rod_charges = [
    {
        "q": 1.0,
        "pos": np.array([-rod_length / 2 + i * rod_length / (num_charges - 1), 0]),
    }
    for i in range(num_charges)
]

x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
Ex, Ey = E_field_point_charges(X, Y, rod_charges)
E = np.sqrt(Ex**2 + Ey**2)
angle = np.arctan2(Ey, Ex)  # range [-pi, pi]

show_field_lines = True
show_field_direction = False
show_field_magnitude = False

# Plotting
fig, ax = plt.subplots()
for charge in rod_charges:
    ax.scatter(
        *charge["pos"],
        c="red" if charge["q"] > 0 else "blue",
        s=50,
        edgecolors="k",
        linewidths=0.5,
    )

title_str = "Electric field from a charged rod: "
if show_field_lines:
    ax.streamplot(X, Y, Ex, Ey, color="k", density=1, linewidth=0.5)
    title_str += "Field Lines"
if show_field_direction:
    ax.imshow(
        angle,
        extent=(x.min(), x.max(), y.min(), y.max()),
        origin="lower",
        cmap="hsv",
    )
    title_str += " with field direction"
elif show_field_magnitude:
    im = ax.imshow(
        E,
        extent=(x.min(), x.max(), y.min(), y.max()),
        origin="lower",
        cmap="inferno",
        norm=LogNorm(vmin=1e-3, vmax=E.max()),
    )
    plt.colorbar(im, ax=ax, label="Field Magnitude")
    title_str += " with field magnitude"
ax.set_aspect("equal")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title(title_str)
plt.show()
```
